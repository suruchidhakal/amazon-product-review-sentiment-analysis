```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Setup global variables once
en_stopwords = stopwords.words('english')
if 'not' in en_stopwords:
    en_stopwords.remove("not")
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove stopwords except 'not'
    text = ' '.join([word for word in text.split() if word not in en_stopwords])
    
    # 3. Replace '*' with 'star' (optional, as you had)
    text = re.sub(r"\*", "star", text)
    
    # 4. Remove punctuation
    text = re.sub(r"[^\w\s]", "", text)
    
    # 5. Tokenize
    tokens = word_tokenize(text)
    
    # 6. Stem tokens
    stemmed_tokens = [ps.stem(token) for token in tokens]
    
    # 7. Lemmatize tokens
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in stemmed_tokens]
    
    # 8. Join back to string
    return " ".join(lemmatized_tokens)
```

```python
data['processed_reviews'] = data['reviews.text'].apply(preprocess_text)
```

**Create Labels**

```python
data['sentiment'] = data['reviews.rating'].apply(lambda x: 1 if x >= 4 else 0)
data[['reviews.text', 'reviews.rating', 'sentiment']].head()
```

**splitting Data**

```python
x = data['reviews.text']
y = data['sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print("Training reviews:", len(X_train))
print("Testing reviews:", len(X_test))
```

**TF_IDF**

```python
tfidf = TfidfVectorizer(max_features=5000)
```

```python
X_train_tfidf = tfidf.fit_transform(X_train)
```

```python
X_test_tfidf = tfidf.transform(X_test)
```

**Training the Model**

```python
model = LogisticRegression()
```

```python
model.fit(X_train_tfidf,y_train)
```

```python
print(type(X_test))
print(X_test[:5])
```

```python
X_test_tfidf = tfidf.transform(X_test)  
y_pred = model.predict(X_test_tfidf) 
```

```python
print("Length of y_test:", len(y_test))
print("Length of y_pred:", len(y_pred))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
